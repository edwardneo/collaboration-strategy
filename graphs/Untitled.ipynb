```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rewards = [-0.1094, -0.1030, -0.0911, -0.0856, -0.0623, -0.0808, -0.1319, -0.1015, -0.0953, -0.0692]

# X-axis values (intervals)
intervals = np.arange(10)

# Plot the rewards over intervals
plt.figure(figsize=(10, 6))
plt.plot(intervals, rewards, marker='o', linestyle='-')

# Set labels and title
plt.xlabel('Epochs')
plt.ylabel('Rewards')
plt.title('Rewards Vs. Epochs')

# Show the plot
plt.grid(True)
plt.savefig('rewards_over_epochs.png')
plt.show()
```

```python
# Arrays of values
bag0_values = [0.702, 1.123, 1.324, 1.156, 0.986, 1.053, 0.972, 0.959, 1.252]
bag1_values = [1.803, 1.527, 1.906, 1.704, 1.838, 1.460, 1.758, 2.442, 2.303]
bag2_values = [0.607, 0.472, 0.635, 0.371, 0.227, 0.402, 1.057, 0.688, 0.569]
bag3_values = [0.919, 0.486, 0.519, 1.022, 1.029, 0.830, 0.188, 0.130, 0.263]

# X-axis values (intervals)
intervals = np.arange(9)

# Plot the values on one graph
plt.figure(figsize=(10, 6))
plt.plot(intervals, bag0_values, label='bag[0]')
plt.plot(intervals, bag1_values, label='bag[1]')
plt.plot(intervals, bag2_values, label='bag[2]')
plt.plot(intervals, bag3_values, label='bag[3]')

# Set labels and title
plt.xlabel('Epochs')
plt.ylabel('Number in Bag')
plt.title('Bag Array Vs. Epochs')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.savefig('bag_array_over_epochs.png')
plt.show()
```
